```python
#importing libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn import linear_model
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

# Part 1- Loading the Dataset

```python
#using pandas to read the dataset
slime_data = pd.read_csv("https://raw.githubusercontent.com/profmcnich/example_notebook/main/science_data_large.csv")
slime_data.head(10) #printing the first 10 rows of the csv dataset
```

# Part 2 - Splitting Dataset

```python
#x-axis represents the featurs and y-axis represnts the label
x_axis = slime_data[['Temperature °C', 'Mols KCL']].to_numpy()
y_axis = slime_data['Size nm^3'].to_numpy()
# Split the features and labels into a training/test set in which 90% train and 10% test using sklearn module
X_train, X_test, Y_train, Y_test = train_test_split(x_axis, y_axis, train_size=.90, random_state=0)
X_train, X_test, Y_train, Y_test # displaying split data
```

# Part 3 - Linear Regression

```python
# We can use sklearn to train a model on the training set
Linear_reg = linear_model.LinearRegression()
Linear_reg.fit(X_train, Y_train)
# Predicting the output of that sample with the trained model by creating a sample datapoint
print(Linear_reg.predict(X_train))
```

```python
# score for that model
print("Score:" ,Linear_reg.score(X_train,Y_train))
```

```python
# coefficents and intercept from the model
print("Co-efficents: ",Linear_reg.coef_)
print("Intercept: ",Linear_reg.intercept_)
```

Based on the score of the model, it is not fully accurate because it is a 2D matrix even tho the data is linear.


# Part 4 - Cross Validation

```python
#We can use cross_val_score function to repeat the experiment across many shuffles of the data. In this case we did 15 shuffels
cross_scores= cross_val_score(Linear_reg, X_test, Y_test, cv=15)
cross_scores.mean()
```

The mean of the cross val scores after 15 shuffles is 83 percent, indicating that it is attempting to approximate the true value of Linear Regression.

```python
PolyReg = PolynomialFeatures(degree=2)
PolyReg_train = PolyReg.fit_transform(X_train)
PolyReg_test = PolyReg.fit_transform(X_test)

BayReg = linear_model.BayesianRidge().fit(PolyReg_train, Y_train)
BayReg.score(PolyReg_test, Y_test) # 1.0 meaning 100% accuracy
```

```python
# Using the PolynomialFeatures library perform another regression on an augmented dataset of degree 2
polyReg = PolynomialFeatures(degree=2)
X_train = polyReg.fit_transform(X_train)
X_test = polyReg.fit_transform(X_test)
model = linear_model.LinearRegression()
model.fit(X_train, Y_train)
```

```python
#repeating what we did for part 3. Returning the score, coeffeicent, intercepts and prediction of the model
print("Score:",model.score(X_train, Y_train))
print("Coeffiencent:",model.coef_)
print("Intercepts:",model.intercept_)
print("Prediction:",model.predict(X_test))
```

This shows we have polynomial distribution because the score is 100%.
